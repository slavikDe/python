```python
text_file = open('./text2.txt', 'r')
text = text_file.read()
text_file.close()
text
```

```python
rEmail = re.compile(r'[\w\.-]+@[\w\.-]+')
emails = rEmail.findall(text)
emails
```

```python
rEmail = re.compile(r'[\w\.-]+@[\w\.-]+')
emails = rEmail.findall(text)
emails

newEmails = [re.compile(r'\W').sub('-', email) for email in emails]

for old_email, new_email in zip(emails, newEmails):
    text = text.replace(old_email, new_email)

text
```

```python
from nltk.tokenize import regexp_tokenize, sent_tokenize


from nltk.stem.lancaster import LancasterStemmer
from nltk.stem.snowball import EnglishStemmer
```

```python
text_file = open('./text2.txt', 'r')
text = text_file.read()
text = text[1:]
text_file.close()
```

## а) провести токенізацію слів, видаливши поштові адреси 

```python

# pattern = r'\b[\w\.-]+@[\w\.-]+\.\w+\b'  

# tokens = regexp_tokenize(text, pattern=pattern, gaps=True)
# tokens
```

```python
from nltk.tokenize import word_tokenize

clean_text = re.sub(r'[\w\.-]+@[\w\.-]+', '', text)

tokens = word_tokenize(clean_text)

tokens
```

## б) Порахувати кількість слів в передостанньому реченні (не враховуючи знаки пунктуації та інші спеціальні символи); 

```python
sent = sent_tokenize(text)[-2]
tokens = regexp_tokenize(sent, r'\w+')

print(sent)
print(len(tokens))
```

##  в) позначити частини мови, вивести усі іменники; 

```python
tokens = regexp_tokenize(text, r'\w+')

pos_tagged = nltk.pos_tag(tokens)

nouns = [word for word, pos in pos_tagged if pos in ['NN', 'NNS', 'NNP', 'NNPS']]
nouns
```

## г) провести пошук кореня слів дев’ятого речення.

```python

ninethSent = sent_tokenize(text)[8]
tokens = regexp_tokenize(ninethSent, pattern=r'\w+')
```

```python
lst = LancasterStemmer()
result = []

for word in tokens:
    result.append(lst.stem(word))
    
lancasterStemmer = " ".join(result)
```

```python
sst = EnglishStemmer()
result = []

for word in tokens:
    result.append(sst.stem(word))

englishStemmer = " ".join(result)
```

```python
print("Original text: ", sent_tokenize(text)[8])
print("Lancaster Stemmer: ", lancasterStemmer)
print("English Stemmer: ", englishStemmer)
```
